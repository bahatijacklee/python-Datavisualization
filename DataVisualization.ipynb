```python
# Importing necessary libraries
import matplotlib.pyplot as plt
import numpy as np

# Data from the table
names = ["Abraham Lincoln", "John Kennedy", "David Beckham", "Jeremy David", "Enock Ariel", "Delmar Javier"]
english = [97, 87, 89, 83, 85, 93]
database = [78, 87, 83, 81, 97, 76]
calculus = [75, 89, 70, 85, 89, 79]
computer_support = [73, 76, 77, 87, 99, 87]
ai = [87, 83, 82, 89, 76, 81]
total = [410, 422, 401, 425, 446, 416]


# Bar chart visualization
plt.figure(figsize=(10,6))
barWidth = 0.15

# Set position of bar on X axis
r1 = np.arange(len(names))
r2 = [x + barWidth for x in r1]
r3 = [x + barWidth for x in r2]
r4 = [x + barWidth for x in r3]
r5 = [x + barWidth for x in r4]

# Make the plot
plt.bar(r1, english, color='blue', width=barWidth, edgecolor='grey', label='English')
plt.bar(r2, database, color='green', width=barWidth, edgecolor='grey', label='Database')
plt.bar(r3, calculus, color='red', width=barWidth, edgecolor='grey', label='Calculus')
plt.bar(r4, computer_support, color='purple', width=barWidth, edgecolor='grey', label='Computer Support')
plt.bar(r5, ai, color='orange', width=barWidth, edgecolor='grey', label='AI')

# Add labels
plt.xlabel('Names', fontweight='bold')
plt.xticks([r + barWidth for r in range(len(names))], names, rotation=45)
plt.ylabel('Scores')
plt.title('Scores in Different Subjects by Student')

# Create legend & Show graphic
plt.legend(loc='upper left', bbox_to_anchor=(1, 1))  # positions the legend box outside the upper-left corner of the axes.
plt.tight_layout()
plt.show()

# Line graph visualization
plt.figure(figsize=(10,6))
for i, subject in enumerate([english, database, calculus, computer_support, ai]):
    plt.plot(names, subject, marker='v', label=['English', 'Database', 'Calculus', 'Computer Support', 'AI'][i])

# Add labels
plt.xlabel('Names')
plt.ylabel('Scores')
plt.title('Line Graph of Scores by Subject')
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show()

# Pie chart visualization of total scores
plt.figure(figsize=(8,8))
plt.pie(total, labels=names, autopct='%1.1f%%', startangle=140, colors=['blue', 'green', 'red', 'purple', 'orange', 'yellow'])
plt.title('Total Scores Distribution among Students')
plt.tight_layout()
plt.show()
```
